# Seleções de frequência

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=';')
dados
```

# Selecionar imóveis classificados como apartamento

```python
selecao = dados['Tipo'] == 'Apartamento'
selecao
```

```python
apartamentos = dados[selecao]
apartamentos.index = range(apartamentos.shape[0])
apartamentos
```

```python
apartamentos.shape[0]
```

# Selecionar imóveis classificados como 'Casa', 'Casa de Condomínio', 'Casa de Vila'  

```python
selecao = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
selecao
```

```python
casas = dados[selecao]
casas.index = range(casas.shape[0])
casas
```

# Selecionar imóveis entre 60 e 100 m2

```python
selecao = (dados['Area'] >= 60) & (dados['Area'] <= 100)
selecao
```

```python
imoveis_60_100 = dados[selecao]
imoveis_60_100.index = range(imoveis_60_100.shape[0])
imoveis_60_100
```

# Selecionar imóveis com pelo menos 4 quartos e aluguel menor que 2000,00

```python
selecao = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
selecao
```

```python
imoveis_4quartos_aluguelAte2000 = dados[selecao]
imoveis_4quartos_aluguelAte2000.index = range(imoveis_4quartos_aluguelAte2000.shape[0])
imoveis_4quartos_aluguelAte2000
```
