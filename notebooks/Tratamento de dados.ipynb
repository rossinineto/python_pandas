# Tratamento dos dados

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=';')
dados.head(20)
```

```python
dados.isnull()
```

```python
dados.notnull()
```

```python
# valor, condominio e iptu com alguns registros nulos
dados.info()
```

```python
dados[dados['Valor'].isnull()]
```

```python
dados.dropna(subset=['Valor'], inplace=True)
dados[dados['Valor'].isnull()]
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados['Tipo'].drop_duplicates()
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
dados[selecao]
```

```python
# negativa da seleção passando o resultado para a própria variável
# dados agora fica sem apartamentos com condomínio nulo
dados = dados[~selecao]
dados[selecao]
```

```python
print(dados[dados['Condominio'].isnull()].shape[0])
print(dados[dados['IPTU'].isnull()].shape[0])
```

```python
# preenchendo dados faltantes com '0'
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```

```python
print(dados[dados['Condominio'].isnull()].shape[0])
print(dados[dados['IPTU'].isnull()].shape[0])
```

```python
dados.info()
```

```python
dados.to_csv('dados/aluguel_residencial_tratado.csv', sep=';', index=False)
```
