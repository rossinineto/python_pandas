```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
dados.head(20)
```

```python
classes_quartos = [0, 2, 4, 6, 100]
quartos = pd.cut(dados['Quartos'], classes_quartos)
quartos
```

```python
pd.value_counts(quartos)
```

```python
labels = ['1 a 2 quartos', '3 a 4 quartos', '5 a 6 quartos', '7 quartos ou mais']
```

```python
quartos = pd.cut(dados['Quartos'], classes_quartos, labels=labels)
pd.value_counts(quartos)
```

```python
# para ver que há registros de imóveis sem quartos
labels = ['0 a 2 quartos', '3 a 4 quartos', '5 a 6 quartos', '7 quartos ou mais']
quartos = pd.cut(dados['Quartos'], classes_quartos, labels=labels, include_lowest=True)
pd.value_counts(quartos)
```
