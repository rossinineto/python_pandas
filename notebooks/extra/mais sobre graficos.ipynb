```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 8))
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
dados.head(10)
```

```python
area = plt.figure()
```

```python
grafico_1 = area.add_subplot(2, 2, 1)
grafico_2 = area.add_subplot(2, 2, 2)
grafico_3 = area.add_subplot(2, 2, 3)
grafico_4 = area.add_subplot(2, 2, 4)
```

```python
grafico_1.scatter(dados['Valor'], dados['Area'])
grafico_1.set_title('Valor X Area')

grafico_2.hist(dados['Valor'])
grafico_2.set_title('Valor')

dados_grafico_3 = dados['Valor'].sample(100)
dados_grafico_3.index = range(dados_grafico_3.shape[0])
grafico_3.plot(dados_grafico_3)
grafico_3.set_title('Amostra')

grupos = dados.groupby('Tipo')['Valor']
label = grupos.mean().index
valores = grupos.mean().values
grafico_4.bar(label, valores)
grafico_4.set_title('Valor médio por tipo')
```

```python
dados = pd.read_csv('dados/aluguel_amostra.csv', sep=';')
dados.head(10)
```

```python
area = plt.figure()

g1 = area.add_subplot(1, 2, 1)
g2 = area.add_subplot(1, 2, 2)

grupo1 = dados.groupby('Tipo Agregado')['Valor']
label = grupo1.count().index
valores = grupo1.count().values
g1.pie(valores, labels = label, autopct='%1.1f%%')
g1.set_title('Total de Imóveis por Tipo Agregado')

grupo2 = dados.groupby('Tipo')['Valor']
label = grupo2.count().index
valores = grupo2.count().values
g2.pie(valores, labels = label, autopct='%1.1f%%', explode = (.1, .1, .1, .1, .1))
g2.set_title('Total de Imóveis por Tipo')
```
