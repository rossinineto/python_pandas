# Imóveis Residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
dados.head(20)
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencia = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
```

```python
residencia
```

```python
selecao = dados['Tipo'].isin(residencia)
selecao.head(20)
```

```python
dados_residencial = dados[selecao]
dados_residencial.head(20)
```

```python
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
dados_residencial.shape[0]
```

```python
dados.shape[0]
```

```python
# reconstruindo index
dados_residencial.index = range(dados_residencial.shape[0])
dados_residencial.head(20)
```

# Exportando os dados

```python
# o cuidado para não exportar com index
dados_residencial.to_csv('dados/aluguel_residencial.csv', sep=';', index=False)
```
