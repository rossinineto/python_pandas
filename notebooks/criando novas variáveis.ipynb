```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial_tratado.csv', sep=';')
dados.head(20)
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head(20)
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
dados.head(20)
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
dados.head(20)
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
dados.head(20)
```

```python
dados['Tipo'].drop_duplicates()
```

```python
tipo_casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
# a função tem que ter o nome 'lambda'
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in tipo_casa else 'Apartamento')
```

```python
dados.head(20)
```

## excluindo variáveis

```python
dados_aux = pd.DataFrame(dados[:])
dados_aux.head(20)
```

```python
# primeira forma
del dados_aux['Valor Bruto']
dados_aux.head(20)
```

```python
# segunda forma
dados_aux.pop('Valor Bruto m2')
dados_aux.head(20)
```

```python
# terceira forma
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis=1, inplace=True)
dados.head(20)
```

```python
dados.to_csv('dados/aluguel_residencial_tratado_agregado.csv', sep=';', index=False)
```
